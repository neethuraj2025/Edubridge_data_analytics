## Countplot

```python
print(d['Age'].value_counts())
plt.figure(figsize=(18,10))
sns.countplot(x=d['Age'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(d['Pregnancies'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d['Pregnancies'])
plt.xticks(rotation=90)
plt.show()
```

# Pie Diagram

```python
#Add figuresize
plt.title('Pregnancies',fontsize=20)
d['Pregnancies'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True, figsize=(25,25))
plt.show()
```

## Heat map

```python
n_data=['BMI','Pregnancies','Age','DiabetesPedigreeFunction']
plt.figure(figsize=(10,5))
sns.heatmap(d[n_data].corr(),annot=True,fmt='.4f',cmap='coolwarm_r',center=0)
plt.show()
```

```python
d.hist(bins=50, figsize=(20,15),color='red')
plt.show()
```

## stripplot

```python
sns.set_style("darkgrid")
sns.stripplot(x="BMI", y="Age", data=d) 
plt.show()
```

## kdeplot

```python
sns.kdeplot(d['BMI'],color='yellow',vertical=True,shade=True)
plt.show()
```
