## Exploring data with queries

### Most expensive car

```python
highestprice = max(d['price'])
d[(d['price']==highestprice)]
```

### Print all toyota car details

```python
d[(d['company']=='toyota')]
```

### Count total cars per company

```python
d['company'].value_counts()
```

### Find each company's highest price car

```python
company_name = d.groupby('company')
price_cw = company_name['company','price'].max()
price_cw
```

### Average mileage of each car making company

```python
company_name = d.groupby('company')
avg_m_c = company_name['company','average-mileage'].mean()
avg_m_c
```

### Sort by price column

```python
d.sort_values(by=['price'])
```

## Data visualization

```python
plt.bar(d['horsepower'].sample(10),d['average-mileage'].sample(10),width=15)
plt.title('Horsepower Vs Average-mileage',fontsize=15,color='red')
plt.xlabel('Horsepower',fontsize=15)
plt.ylabel('average-mileage',fontsize=15)
plt.grid(True)
plt.show()
```

```python
print(d['company'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d['company'])
plt.xticks(rotation=90)
plt.show()
```

```python
print(d['num-of-cylinders'].value_counts())
plt.figure(figsize=(10,6))
sns.countplot(x=d['num-of-cylinders'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.bar(d['num-of-cylinders'],d['horsepower'],color="red")
#plt.title('Profit Vs Quantity',fontsize=15,color='red')
plt.title('num-of-cylinders Vs horsepower',fontsize=15,color='blue')
plt.xlabel('num-of-cylinders',fontsize=15)
plt.ylabel('horsepower',fontsize=15)
plt.grid(True)
plt.show()
```

```python
sns.relplot(x='company',y='price',hue='company',data=d)
#plt.figure(figsize=(10,50))
plt.show()
```

```python
plt.title('company',fontsize=20)
d['company'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(10,10))
plt.show()
```

```python
plt.figure(figsize=(5,8))
sns.countplot(x="body-style",data=d,palette='rocket_r',order=d['body-style'].value_counts().index)
plt.xlabel('Body-style',fontsize=25)
plt.xticks(rotation=70)
plt.show()
```
